```python
import pandas as pd
import os
import csv

#Update these variables when new data comes in/path has changed
path=r'C:\Users\brnma\OneDrive\Desktop\Research\RF MSU Data\Feature Sets\Burst '
numBursts = 5
numRegions = 20
numDevices = 6


firstFile = True
firstBurst = True
currentIndex = 0
df = pd.DataFrame(index = range(0, numDevices*numRegions), columns=['Burst', 'Region', 'Device'])
maindf = pd.DataFrame()

   
def handleFirstFile(file, tempPath):
    global firstFile
    global df
    firstFile = False  

    j = 0
    
    
    fileDataFrame = pd.read_csv(tempPath + "\\" + file)
    for col in fileDataFrame.columns:
        k = 0
        while k < numRegions:
            df['Device'][j] = col
            df['Region'][j] = k
            df['Burst'][j] = ((j // numRegions)% numBursts) + 1
            j+=1
            k+=1
    
def enterValues(file, tempPath, burst):
    #global currentIndex
    global df
    i = 0
    fileDataFrame = pd.read_csv(tempPath + "\\" + file)
    
    if firstBurst:
        df[file[:-4]] = 'NaN'#Create column headings and set default values of columns to NaN

    while i < numRegions:
        for col in fileDataFrame.columns:     
            column = findColumn(file)
            df[column][i + ((burst-1) * numRegions)] = fileDataFrame[col][i]
#             i+=((burst-1) * numRegions)
        i+=1
    

def findColumn(file):
    i = 0
    column = ""
    while(i < len(file)):
        if file[i].isalpha():
            column += file[i]
            i+=1
        else:
            break
    return column
        
    
    

def loop(path):
    global df
    global firstBurst
    burst = 0
    
    while burst < numBursts:
        burst+=1
        tempPath = path + str(burst)
        
        
        if not os.path.exists(tempPath):
            print("Error: The Pathway " + tempPath +  " does not exist")
            break
            
        allFiles = [f for f in os.listdir(tempPath) if os.path.isfile(os.path.join(tempPath, f))] #From https://stackoverflow.com/questions/11968976/list-files-only-in-the-current-directory
        for file in allFiles:    
            if(file.endswith(".csv")) and firstFile:
                handleFirstFile(file, tempPath)
                
            if(file.endswith(".csv")):
                enterValues(file, tempPath, burst)
        
        firstBurst = False
    
            
                
                
                
                
                
                
loop(path)
print(df.head(120))
```
