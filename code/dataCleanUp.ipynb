```python
#THaW Project
#This program is meant to help create a single excel file from the data collected by MSU
#6/12/2019
#Code Written By: Manzi Bryan


import pandas as pd
import os
import csv

#If this is your first time running this program, after downloading the data from the shared google drive
#Enter the path where you downloaded the file
#Update the numBursts variable below with the number of bursts measured(same as the number of folders in the features set)
#Install necessary packages (Use miniconda!)
#Hit 'Run' and then search for a file called msudata.xlsx


#Change these globals depending on the data set
path=r'C:\Users\brnma\OneDrive\Desktop\Research\THAW-rf\RF MSU Data\Feature Sets\Burst '
numBursts = 5

#Some necessary globals: DO NOT CHANGE

firstFile = True
firstBurst = True
currentIndex = 0


#Builds the shell of the table using information obtained from the first file
def handleFirstFile(file, tempPath):
    global firstFile
    global df
    global numDevices
    global numRegions
    
    firstFile = False    
    j = 0
    fileDataFrame = pd.read_csv(tempPath + "\\" + file)
    numDevices = len(fileDataFrame.columns)
    numRegions = len(fileDataFrame.index) - 1
    df = pd.DataFrame(index = range(0, numDevices*numBursts*numRegions), columns=['Burst', 'SubRegion', 'Device'])
    
    for col in fileDataFrame.columns:
        k = 0
        while k < numRegions * numBursts:
            df['Device'][j] = col
            df['SubRegion'][j] = k % numRegions
            df['Burst'][j] = ((j // numRegions)% numBursts) + 1
            j+=1
            k+=1
#     print(fileDataFrame)
    print(df)
            
#dynamically adds to the table as it reads from files 'enters values' into the table  
def enterValues(file, tempPath, burst):
    global df
    
    column = findColumn(file)
    fileDataFrame = pd.read_csv(tempPath + "\\" + file)
    
    if firstBurst:
        df[file[:-4]] = 'NaaN'#Create column headings and set default values of columns to NaN
        
    i = (burst-1)*numRegions
    j = i
    while i < j + numRegions:
        
        for col in fileDataFrame.columns:
            df[column][i] = fileDataFrame[col][i % numRegions]
            i += numBursts * numRegions
        i -= numBursts * numRegions * numDevices
        i +=1
    

#Finds the right column to place a file into
def findColumn(file):
    i = 0
    column = ""
    while(i < len(file)):
        if file[i].isalpha():
            column += file[i]
            i+=1
        else:
            break
    return column
        
    
    
#the main function, loops through all of the folders in the directory, calls relevant functions
def loop(path):
    global df
    global firstBurst
    burst = 0
    
    while burst < numBursts:#for all of the burst folders
        burst+=1
        tempPath = path + str(burst)#make a path to that folder
        
        
        if not os.path.exists(tempPath):
            print("Error: The Pathway " + tempPath +  " does not exist")
            break
            
        allFiles = [f for f in os.listdir(tempPath) if os.path.isfile(os.path.join(tempPath, f))] #list of all of the files in said directory
        for file in allFiles:    
            if(file.endswith(".csv")) and firstFile:#first file is going to be used to format the dataframe
                handleFirstFile(file, tempPath)
                
            if(file.endswith(".csv")):#enter values places the features into the dataframe
                enterValues(file, tempPath, burst)
        
        firstBurst = False           
            
                
                
                
                
                
                
loop(path)

df.to_excel(r'msudata2.xlsx')  # doctest: +SKIP
df.to_csv(r'msudata2.csv')  # doctest: +SKIP
```
