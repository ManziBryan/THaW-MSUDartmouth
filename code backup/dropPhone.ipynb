```python
import pandas as pd
import pickle

tsPath = r'C:\Users\brnma\TSFeatures\TSNoPyleorFever.xlsx'
ts = pd.read_excel(tsPath)

def dropPhone(path):
    global ts
    
    clf =  r'C:\Users\brnma\RandomForest d=22 n_est=55Model.sav'
    
    loadedModel = pickle.load(open(clf, 'rb'))
    
    contaminated = pd.read_excel(path)
    
    devices = ts['Device'].unique()
    
    contaminatedNoDev = contaminated.drop(columns= ['Device'])
    contaminatedNoDev.drop(contaminatedNoDev.columns[contaminatedNoDev.columns.str.contains
                                                     ('Unname',case = False)],axis = 1, inplace = True)
    
    pred = loadedModel.predict(contaminatedNoDev)
    pred_proba = loadedModel.predict_proba(contaminatedNoDev)
    print(pred)
    
    i = 0
    while i < len(pred):
        if devices[pred[i]] != 'Gian Iphone':
            ts = ts.append(contaminated.loc[i, : ])
        else:
            print(devices[pred[i]] + ': ' + str(pred_proba[i].max()))
        i+=1
    
if __name__ == '__main__':
    contaminatedPaths = [r'C:\Users\brnma\TSFeatures\legalPyle1.xlsx', r'C:\Users\brnma\TSFeatures\legalPyle2.xlsx',
                        r'C:\Users\brnma\TSFeatures\legalFever1.xlsx', r'C:\Users\brnma\TSFeatures\legalFever2.xlsx']
    
    
    for path in contaminatedPaths:
        dropPhone(path)
    ts.to_excel('uncontaminated.xlsx')
```
