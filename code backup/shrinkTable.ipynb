```python
import pandas as pd

if __name__ == '__main__':
    target = 'Device'
    table = pd.read_excel('table.xlsx')
    devices = table[target].unique()
    smallerTable = pd.DataFrame(columns = table.columns)
    i = 0
    for device in devices:
        smallerTable = smallerTable.append(table[table.Device == device].mean(), ignore_index=True)
        smallerTable['Std Conf'][i] = table[table.Device == device]['Avg Conf'].std()
        i+=1
    smallerTable[target] = devices
    print(smallerTable)
    smallerTable.to_excel("smallerTable.xlsx")
    
```
