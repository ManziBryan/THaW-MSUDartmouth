```python

# This Program takes a directory with many .mat files and converts the files to excel files 
# with the right naming format

# Manzi Bryan, Summer 2019
# KotzGroup


from os import listdir
import os
from os.path import isfile, join
import numpy as np
import pandas as pd
import tables
from scipy.io import loadmat
import shutil


# Global Variables

path= r'.\Replays'
saveTo = r'HackRf'

subdirs = [f.path for f in os.scandir(path) if f.is_dir() ]

def getDeviceName(subdir):
    i = -1
    device = ''
    while abs(i) < len(subdir):
        if not subdir[i] == '\\':
            device += subdir[i]
            i -= 1
        else:
            return device[::-1]
def correctNumZeros(dp, i):
    i = str(i)
    string = ''
    while len(string) < dp-len(i):
        string += '0'
    string += i
    return string

if __name__ == '__main__':
    
    if os.path.isdir(saveTo):
        shutil.rmtree(saveTo)
        
    os.mkdir(saveTo)
   
    for subdir in subdirs:
        device = getDeviceName(subdir)
        
        
        os.mkdir( saveTo + '\\' + device)
        
        files = [f for f in listdir(subdir) if isfile(join(subdir, f))]
        i = 0
        
        for file in files:
            if file.endswith('.MAT'):
                file = subdir +  '\\' + file
                f = loadmat(file)
                print(file)
                
                if 'Y' in f.keys():
                    df = pd.DataFrame(f['Y'])
                    df.to_excel(saveTo + '\\' + device + '\\Burst'  + correctNumZeros(3, i) + '.xlsx')  # doctest: +SKIP
                    i+=1
            
           
        
   
```
